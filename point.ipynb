```python
def bar_chart(feature):
    male = cus[cus['CLNT_GENDER']=='M'].feature.value_counts()
    female = cus[cus['CLNT_GENDER']=='F'].feature.value_counts()
    df = pd.DataFrame([male,female])
    df.index = ['male','female']
    df.plot(kind='bar',stacked=True, figsize=(10,5))
```

```python
m=cus[cus['CLNT_GENDER']=='M'].CLNT_AGE.value_counts()
f=cus[cus['CLNT_GENDER']=='F'].CLNT_AGE.value_counts()
mf=pd.DataFrame([m,f])
mf.index=['male','female']
mf.plot(kind='bar',stacked=True)
```
